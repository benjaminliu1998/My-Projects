```python
lactation_string = open("/content/drive/My Drive/#7 Fall 2019 Senior/CSC Honor Research/Pediatrics/Saved Data New/X_train_all_in_one_string_lactation").read()

!pip install python-rake==1.4.4
import RAKE
import operator

# Reka setup with stopword directory
stop_dir = "/content/drive/My Drive/#7 Fall 2019 Senior/CSC Honor Research/Pediatrics/SmartStoplist2.txt"
rake_object = RAKE.Rake(stop_dir)

# Extract keywords in Pedia
keywords = rake_object.run(lactation_string)
print("Lac keywords: ", keywords)

with open("/content/drive/My Drive/#7 Fall 2019 Senior/CSC Honor Research/Pediatrics/Saved Data New/Keywords_Lac.txt", 'w') as f: 
  for elmt in keywords:
    if len(elmt[0].split()) == 2: # length 2
      print("Length 2: " + elmt[0] + " Score: " + str(elmt[1]))
      f.write("Length 2: " + elmt[0] + " Score: " + str(elmt[1])+ '\n')
    elif len(elmt[0].split()) == 1: # length 1
      print("Length 1: " + elmt[0] + " Score: " + str(elmt[1]))
      f.write("Length 2: " + elmt[0] + " Score: " + str(elmt[1])+ '\n')
  f.write("Lac keywords: " + str(keywords))
f.close()
```
