```python
### This file extracts text from Geriartic section, lactation section, pregnancy section from all SPL files. 

from lxml import etree
import os, random
import json

DATA_PATH = "/content/drive/My Drive/#7 Fall 2019 Senior/CSC Honor Research/Pediatrics/All xml Files"
os.chdir(DATA_PATH)
#!ls -al


'''SPL PARSER for other 4 section data-
    1. Extracts section text from ALL files for each section -  
        34082-8	GERIATRIC USE SECTION
        34080-2	NURSING MOTHERS SECTION
        42228-7	PREGNANCY SECTION
        77290-5	LACTATION SECTION
'''

namespaces={"v3":"urn:hl7-org:v3",}

class SPLDocument:
    def __init__(self):
        self.myDict = {}
    
    def get_metadata_4_sections(self, tree, fileName, sec):
        self.myDict['SPLFileName']= fileName
        # Get the drug name
        drugName = tree.xpath(".//v3:genericMedicine/v3:name/text()",namespaces=namespaces)

        if(len(drugName)==0):
          return -1

        self.myDict['genericName']= drugName[0]
        
        # Get the specific section text
        section_text_list = []
        elementText = " "
        section_text = ''

        query = "//v3:code[@code='"+sec+"']"
        sectionTag = tree.xpath(query,namespaces=namespaces)
        elementText = " "
        if len(sectionTag) != 0:
            sectionTag = sectionTag[0]
            for ancestor in sectionTag.iterancestors('{urn:hl7-org:v3}section'):    
                elementText += ("".join(ancestor.itertext())).strip() 
                section_text_list.append(elementText)
                break                
            section_text = ''.join(section_text_list)

        if(section_text!=''):
            os.makedirs('/content/drive/My Drive/#7 Fall 2019 Senior/CSC Honor Research/Pediatrics/output_ger/train/geriatrics', exist_ok = True)
            outputfile = "/content/drive/My Drive/#7 Fall 2019 Senior/CSC Honor Research/Pediatrics/output_ger/train/geriatrics/" + fileName.replace('.xml','.txt')
            f= open(outputfile,"wb+")
            section_text = section_text.strip().replace('  ','')
            f.write(section_text.encode('ascii', 'ignore'))
            f.close()
            
            return 0
        return -1       # If section does not exist

# ## Parsing SPL documents 

filePath = r"/content/drive/My Drive/#7 Fall 2019 Senior/CSC Honor Research/Pediatrics/All xml Files/"
p = etree.XMLParser(huge_tree=True)

sections = {'34082-8':['GERIATRIC', 0],'34080-2':['NURSING', 0],'42228-7':['PREGNANCY', 0],'77290-5':['LACTATION', 0]}

count = 0
for doc in os.listdir(filePath):
    spl = SPLDocument()
    tree = etree.parse(filePath+doc,parser=p)
    spl.get_metadata_4_sections(tree, doc, '34082-8') 
    count += 1
    print(str(count) + " processed")
```
